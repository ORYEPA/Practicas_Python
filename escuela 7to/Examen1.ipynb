```python
def alta_alumno(alumnos):
    id = len(alumnos) + 1
    nombre = input("Ingrese el nombre del alumno: ")
    materia = input("Ingrese la materia: ")
    parcial1 = float(input("Ingrese la calificación del parcial 1: "))
    parcial2 = float(input("Ingrese la calificación del parcial 2: "))
    parcial3 = float(input("Ingrese la calificación del parcial 3: "))
    promedio = (parcial1 + parcial2 + parcial3) / 3
    alumno = {"id": id, "nombre": nombre, "materia": materia, "parcial1": parcial1, "parcial2": parcial2, "parcial3": parcial3, "promedio": promedio}
    alumnos.append(alumno)
    print("Alumno registrado con éxito.")

def baja_alumno(alumnos, nombre):
    alumnos = [alumno for alumno in alumnos if alumno["nombre"] != nombre]
    print(f"Alumno '{nombre}' eliminado.")

def consulta_alumnos(alumnos):
    print("Lista de alumnos registrados:")
    for alumno in alumnos:
        print(f"ID: {alumno['id']}, Nombre: {alumno['nombre']}")

def reporte_todos(alumnos):
    if not alumnos:
        print("No hay alumnos registrados.")
        return

    print(f"{'ID':<5}{'Nombre':<15}{'Materia':<15}{'Parcial 1':<10}{'Parcial 2':<10}{'Parcial 3':<10}{'Promedio':<10}")
    total_promedio = 0
    for alumno in alumnos:
        total_promedio += alumno["promedio"]
        print(f"{alumno['id']:<5}{alumno['nombre']:<15}{alumno['materia']:<15}{alumno['parcial1']:<10}{alumno['parcial2']:<10}{alumno['parcial3']:<10}{alumno['promedio']:<10}")
    
    promedio_general = total_promedio / len(alumnos)
    print(f"{'Promedio general:':<50}{promedio_general:<10}")

def reporte_aprobados_no_aprobados(alumnos):
    if not alumnos:
        print("No hay alumnos registrados.")
        return

    print(f"{'ID':<5}{'Nombre':<15}{'Materia':<15}{'Parcial 1':<10}{'Parcial 2':<10}{'Parcial 3':<10}{'Promedio':<10}")
    total_promedio = 0
    for alumno in alumnos:
        total_promedio += alumno["promedio"]
        promedio = alumno["promedio"]
        if promedio < 70:
            promedio = f"{promedio} (NA)"
        print(f"{alumno['id']:<5}{alumno['nombre']:<15}{alumno['materia']:<15}{alumno['parcial1']:<10}{alumno['parcial2']:<10}{alumno['parcial3']:<10}{promedio:<10}")
    
    promedio_general = total_promedio / len(alumnos)
    print(f"{'Promedio general:':<50}{promedio_general:<10}")


def sumatoria():
    n = int(input("Ingrese el valor de n: "))
    
    if n <= 0:
        print("El valor de n debe ser mayor que cero.")
        return
    
    sumatoria = 0
    print("Secuencia 1/(2n-1):")
    for i in range(1, n + 1):
        term = 1 / (2 * i - 1)
        sumatoria += term
        print(f"1/(2*{i}-1) = {term}")

    print(f"Resultado de la sumatoria: {sumatoria}")

def main():
    alumnos = []

    while True:
        print("\nMenú:")
        print("1. Alumnos")
        print("2. Sumatoria")
        print("3. Salir")
        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            while True:
                print("\nMenú Alumnos:")
                print("1. Alta")
                print("2. Baja")
                print("3. Consulta")
                print("4. Reporte de todos")
                print("5. Reporte de aprobados y no aprobados")
                print("6. Volver al menú principal")
                subopcion = input("Seleccione una opción: ")

                if subopcion == "1":
                    alta_alumno(alumnos)
                elif subopcion == "2":
                    nombre = input("Ingrese el nombre del alumno a eliminar: ")
                    baja_alumno(alumnos, nombre)
                elif subopcion == "3":
                    consulta_alumnos(alumnos)
                elif subopcion == "4":
                    reporte_todos(alumnos)
                elif subopcion == "5":
                    reporte_aprobados_no_aprobados(alumnos)
                elif subopcion == "6":
                    break
                else:
                    print("Opción no válida.")
        elif opcion == "2":
            sumatoria()
        elif opcion == "3":
            print("¡Adiós!")
            break
        else:
            print("Opción no válida.")

if __name__ == "__main__":
    main()
```
