```python
#2 tarea
# Lista inicial de artículos
articulos = []

# Función para calcular el costo de producción
def calcular_costo_produccion(articulo):
    cantidad = articulo["Cantidad"]
    factor_materiales = articulo["Factor_Materiales"]
    costo_fijo = articulo["Costo_Fijo"]
    costo_produccion = cantidad * factor_materiales + costo_fijo
    return costo_produccion

# Función para agregar un nuevo artículo
def agregar_articulo():
    nombre = input("Ingrese el nombre del artículo: ")
    descripcion = input("Ingrese la descripción del artículo: ")
    cantidad = float(input("Ingrese la cantidad de unidades producidas: "))
    factor_materiales = float(input("Ingrese el factor de costo de materiales: "))
    costo_fijo = float(input("Ingrese el costo fijo: "))

    articulo = {
        "Nombre": nombre,
        "Descripcion": descripcion,
        "Cantidad": cantidad,
        "Factor_Materiales": factor_materiales,
        "Costo_Fijo": costo_fijo
    }

    articulos.append(articulo)
    print(f"El artículo '{nombre}' ha sido agregado correctamente.")
```

```python
while True:
    print("\n1. Agregar nuevo artículo")
    print("2. Generar informe de Costos de Producción")2
    print("3. Salir")

    opcion = input("Seleccione una opción: ")

    if opcion == "1":
        agregar_articulo()
    elif opcion == "2":
        print("\nInforme de Costos de Producción\n")
        print("{:<30} {:<20} {:<15} {:<15} {:<20}".format("Nombre del Artículo", "Unidades Producidas", "Factor de Costo", "Costo Fijo", "Costo de Producción"))
        print("="*100)

        total_unidades_producidas = 0
        total_factor_costo = 0
        total_costo_fijo = 0
        total_costo_produccion = 0

        for articulo in articulos:
            nombre = articulo["Nombre"]
            unidades_producidas = articulo["Cantidad"]
            factor_costo = articulo["Factor_Materiales"]
            costo_fijo = articulo["Costo_Fijo"]
            costo_produccion = calcular_costo_produccion(articulo)

            total_unidades_producidas += unidades_producidas
            total_factor_costo += factor_costo
            total_costo_fijo += costo_fijo
            total_costo_produccion += costo_produccion

            print("{:<30} {:<20} {:<15} {:<15} {:<20}".format(nombre, unidades_producidas, factor_costo, costo_fijo, costo_produccion))
        
        print("-" * 100)
        print("{:<30} {:<20} {:<15} {:<15} {:<20}".format("Total:", total_unidades_producidas, total_factor_costo, total_costo_fijo, total_costo_produccion))
    elif opcion == "3":
        break
    else:
        print("Opción no válida. Por favor, seleccione una opción válida.")
```
