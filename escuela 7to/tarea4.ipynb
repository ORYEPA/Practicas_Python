```python
# Lista inicial de artículos
articulos = []

# Función para agregar un nuevo artículo
def agregar_articulo():
    nombre = input("Ingrese el nombre del artículo: ")
    descripcion = input("Ingrese la descripción del artículo: ")
    cantidad = int(input("Ingrese la cantidad de artículos: "))
    precio_unitario = float(input("Ingrese el precio unitario de venta: "))

    articulo = {
        "Nombre": nombre,
        "Descripcion": descripcion,
        "Cantidad": cantidad,
        "Precio_Unitario": precio_unitario
    }

    articulos.append(articulo)
    print(f"El artículo '{nombre}' ha sido agregado correctamente.")

# Función para calcular el subtotal
def calcular_subtotal(articulos):
    subtotal = sum(articulo["Cantidad"] * articulo["Precio_Unitario"] for articulo in articulos)
    return subtotal

# Función para calcular el impuesto
def calcular_impuesto(subtotal):
    impuesto = 0.15 * subtotal
    return impuesto

# Función para calcular el total
def calcular_total(subtotal, impuesto):
    total = subtotal + impuesto
    return total

# Generar el informe
cliente = input("Ingrese el nombre del cliente: ")

while True:
    print("\n1. Agregar nuevo artículo")
    print("2. Generar informe de Costos y Precios de Venta")
    print("3. Salir")

    opcion = input("Seleccione una opción: ")

    if opcion == "1":
        agregar_articulo()
    elif opcion == "2":
        print("\nInforme de Costos y Precios de Venta para", cliente, "\n")
        print("{:<30} {:<20} {:<20} {:<20} {:<20}".format("Nombre del Artículo", "Descripción", "Cantidad", "Precio Unitario", "Precio Total"))
        print("="*130)

        for articulo in articulos:
            nombre = articulo["Nombre"]
            descripcion = articulo["Descripcion"]
            cantidad = articulo["Cantidad"]
            precio_unitario = articulo["Precio_Unitario"]
            precio_total = cantidad * precio_unitario

            print("{:<30} {:<20} {:<20} {:<20} {:<20}".format(nombre, descripcion, cantidad, precio_unitario, precio_total))

        subtotal = calcular_subtotal(articulos)
        impuesto = calcular_impuesto(subtotal)
        total = calcular_total(subtotal, impuesto)

        print("="*130)
        print("{:<90} {:<20}".format("Subtotal:", subtotal))
        print("{:<90} {:<20}".format("Impuesto (15%):", impuesto))
        print("{:<90} {:<20}".format("Total:", total))
    elif opcion == "3":
        break
    else:
        print("Opción no válida. Por favor, seleccione una opción válida.")
```
