```python
global alumnos  


def alta_alumno(alumnos):
    id_alumno = len(alumnos) + 1
    nombre = input("Ingrese el nombre del alumno: ")
    nombres = [a['nombre'] for a in alumnos]
    if nombre in nombres:
        print("El nombre del alumno ya existe. No se puede dar de alta.")
        return
    materias = []  # List to store multiple subjects

    while True:
        materia = input("Ingrese la materia (o presione Enter para finalizar): ")
        if not materia:
            break
        materias.append(materia)

    calificaciones = {}
    for materia in materias:
        parcial1 = float(input(f"Ingrese la calificación del parcial 1 para {materia}: "))
        parcial2 = float(input(f"Ingrese la calificación del parcial 2 para {materia}: "))
        parcial3 = float(input(f"Ingrese la calificación del parcial 3 para {materia}: "))
        promedio = (parcial1 + parcial2 + parcial3) / 3
        calificaciones[materia] = {"parcial1": parcial1, "parcial2": parcial2, "parcial3": parcial3, "promedio": promedio}

    alumno = {"id": id_alumno, "nombre": nombre, "materias": calificaciones}

    alumnos.append(alumno)
    print(f"Alumno registrado con ID: {id_alumno}")
def baja_alumno(alumnos, nombre):
    encontrado = False
    for alumno in alumnos:
        if alumno["nombre"] == nombre:
            alumnos.remove(alumno)
            encontrado = True
            print(f"Alumno '{nombre}' eliminado.")
            break
    if not encontrado:
        print(f"Alumno '{nombre}' no encontrado.")

def consulta_alumnos(alumnos):
    print("Consulta de alumnos:")
    nombre_buscar = input("Ingrese el nombre del alumno a buscar: ")

    encontrado = False
    for alumno in alumnos:
        if alumno['nombre'] == nombre_buscar:
            encontrado = True
            print("\nDatos del alumno:")
            print(f"ID: {alumno['id']}")
            print(f"Nombre: {alumno['nombre']}")
            print("Materias:")
            for materia, calificaciones in alumno['materias'].items():
                print(f"  {materia}:")
                print(f"    Parcial 1: {calificaciones['parcial1']}")
                print(f"    Parcial 2: {calificaciones['parcial2']}")
                print(f"    Parcial 3: {calificaciones['parcial3']}")
                print(f"    Promedio: {calificaciones['promedio']}")

            opcion = input("\n¿Desea agregar o actualizar datos para este alumno? (S/N): ")
            if opcion.lower() == "s":
                materia = input("Ingrese la nueva materia (o presione Enter para mantener la actual): ")
                if materia:
                    alumno['materias'][materia] = {}

                for materia, calificaciones in alumno['materias'].items():
                    parcial1 = input(f"Ingrese la nueva calificación del parcial 1 para {materia} (o presione Enter para mantener la actual): ")
                    if parcial1:
                        calificaciones['parcial1'] = float(parcial1)

                    parcial2 = input(f"Ingrese la nueva calificación del parcial 2 para {materia} (o presione Enter para mantener la actual): ")
                    if parcial2:
                        calificaciones['parcial2'] = float(parcial2)

                    parcial3 = input(f"Ingrese la nueva calificación del parcial 3 para {materia} (o presione Enter para mantener la actual): ")
                    if parcial3:
                        calificaciones['parcial3'] = float(parcial3)

                    calificaciones['promedio'] = (calificaciones['parcial1'] + calificaciones['parcial2'] + calificaciones['parcial3']) / 3

                print("Datos actualizados exitosamente.")
            break

    if not encontrado:
        print(f"Alumno '{nombre_buscar}' no encontrado.")
def reporte_todos(alumnos):
    if not alumnos:
        print("No hay alumnos registrados.")
        return

    print(f"{'ID':<5}{'Nombre':<15}{'Materia':<15}{'Parcial 1':<10}{'Parcial 2':<10}{'Parcial 3':<10}{'Promedio':<10}")
    total_promedio = 0
    for alumno in alumnos:
        for materia, calificaciones in alumno['materias'].items():
            total_promedio += calificaciones["promedio"]
            print(f"{alumno['id']:<5}{alumno['nombre']:<15}{materia:<15}{calificaciones['parcial1']:<10}{calificaciones['parcial2']:<10}{calificaciones['parcial3']:<10}{calificaciones['promedio']:<10}")
    
    promedio_general = total_promedio / len(alumnos)
    print(f"{'Promedio general:':<50}{promedio_general:<10}")

def reporte_aprobados_no_aprobados(alumnos):
    if not alumnos:
        print("No hay alumnos registrados.")
        return

    print(f"{'ID':<5}{'Nombre':<15}{'Materia':<15}{'Parcial 1':<10}{'Parcial 2':<10}{'Parcial 3':<10}{'Promedio':<15}")
    total_promedio = 0
    for alumno in alumnos:
        for materia, calificaciones in alumno['materias'].items():
            total_promedio += calificaciones["promedio"]
            promedio = calificaciones["promedio"]
            if promedio < 70:
                promedio = f"{promedio} (NA)"
            print(f"{alumno['id']:<5}{alumno['nombre']:<15}{materia:<15}{calificaciones['parcial1']:<10}{calificaciones['parcial2']:<10}{calificaciones['parcial3']:<10}{promedio:<15}")
    
    promedio_general = total_promedio / len(alumnos)
    print(f"{'Promedio general:':<60}{promedio_general:<10}")


def sumatoria():
    n = int(input("Ingrese el valor de n: "))
    if n <= 0:
        print("El valor de n debe ser mayor que cero.")
        return
    
    sumatoria = 0
    print("Secuencia 1/(2n-1):")
    for i in range(1, n + 1):
        term = 1 / (2 * i - 1)
        sumatoria += term
        print(f"1/(2*{i}-1) = {term}")

    print(f"Resultado de la sumatoria: {sumatoria}")

def main():
    alumnos = []

    while True:
        print("\nMenú:")
        print("1. Alumnos")
        print("2. Sumatoria")
        print("3. Salir")
        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            while True:
                print("\nMenú Alumnos:")
                print("1. Alta")
                print("2. Baja")
                print("3. Consulta")
                print("4. Reporte de todos")
                print("5. Reporte de aprobados y no aprobados")
                print("6. Volver al menú principal")
                subopcion = input("Seleccione una opción: ")

                if subopcion == "1":
                    alta_alumno(alumnos)
                elif subopcion == "2":
                    nombre = input("Ingrese el nombre del alumno a eliminar: ")
                    baja_alumno(alumnos, nombre)
                elif subopcion == "3":
                    consulta_alumnos(alumnos)
                elif subopcion == "4":
                    reporte_todos(alumnos)
                elif subopcion == "5":
                    reporte_aprobados_no_aprobados(alumnos)
                elif subopcion == "6":
                    break
                else:
                    print("Opción no válida.")
        elif opcion == "2":
            sumatoria()
        elif opcion == "3":
            print("¡Adiós!")
            break
        else:
            print("Opción no válida.")

if __name__ == "__main__":
    main()
```
