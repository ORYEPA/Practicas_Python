```python
import cv2

# Load the pre-trained face detection model
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Load the image you want to detect faces in
image_path = 'D:\Practicas_Python\escuela7to/a.jpg'  # Replace with the path to your image
image = cv2.imread(image_path)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Detect faces in the image
faces = face_cascade.detectMultiScale(gray_image, scaleFactor=1.3, minNeighbors=5, minSize=(30, 30))

# Draw rectangles around detected faces
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Display the image with detected faces
cv2.imshow('Face Detection', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

# Load the pre-trained face detection model
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Load the image
image_path = 'D:\Practicas_Python\escuela7to/a.jpg'  # Replace with the path to your image
image = cv2.imread(image_path)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Detect faces in the image
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

# Draw red circles around detected faces
for (x, y, w, h) in faces:
    center_x = x + w // 2
    center_y = y + h // 2
    radius = max(w, h) // 2
    cv2.circle(image, (center_x, center_y), radius, (0, 0, 255), 2)

# Display the image with highlighted faces
cv2.imshow('Face Detection', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import matplotlib.pyplot as plt

# Cargamos la imagen
image_path = 'D:\Practicas_Python\escuela7to/a.jpg'  # Replace with the path to your image
image = cv2.imread(image_path)

# Cargamos el clasificador de deteccin de caras preentrenado de OpenCV
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Convertimos la imagen a escala de grises
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Detectamos caras en la imagen
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

# Dibujamos un crculo alrededor de cada cara detectada
for (x, y, w, h) in faces:
    cv2.circle(image, (x + w // 2, y + h // 2), int(0.6 * w), (0, 255, 0), 2)

# Mostramos la imagen con los crculos
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()

# Imprimimos el nmero de caras detectadas
print(f"Nmero de caras detectadas: {len(faces)}")
```
