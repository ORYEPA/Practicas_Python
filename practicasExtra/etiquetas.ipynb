```python
import cv2
import numpy as np
from matplotlib import pyplot as plt #pa plotear


# Read the image
figura = cv2.imread('animales.png')

# Convert the image to grayscale
gray = cv2.cvtColor(figura, cv2.COLOR_BGR2GRAY)

# Apply a threshold to segment the image
ret, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY)

# Find the contours in the image
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Draw the contours on a black background
color_seg = np.zeros_like(figura)
for i, contour in enumerate(contours):
    cv2.drawContours(color_seg, contours, i, (0, 0, i*30), -1)

# Display the segmented image
# cv2.imshow('Color Segmentation', color_seg)
# cv2.waitKey(0)
plt.imshow(color_seg)
```

```python
import cv2

# Lee la imagen
imagen = figura

# Convierte la imagen a escala de grises
gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

# Aplica un filtro de suavizado para reducir el ruido
suavizado = cv2.GaussianBlur(gris, (5, 5), 0)

# Detecta los bordes en la imagen suavizada utilizando el algoritmo Canny
bordes = cv2.Canny(suavizado, 30, 150)

# Muestra el contorno de la imagen
# cv2.imshow('Contorno', bordes)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
plt.imshow(bordes,cmap="gray",vmin=0,vmax=255)
```
