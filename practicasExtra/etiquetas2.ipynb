```python
import numpy as np
import cv2

# Read image
figura = cv2.imread('animales.png', cv2.IMREAD_GRAYSCALE)

# Add border of zeros
figura = np.pad(figura, ((1,1),(1,1)), mode='constant', constant_values=0)

# Label connected components
[M, N] = figura.shape
lp = 0
lq = 0
lx = 0
etiq = 0
T = []
for i in range(1, M-1):
    for j in range(1, N-1):
        if figura[i,j] == 255:
            lp = figura[i-1,j]
            lq = figura[i,j-1]
            if lp == 0 and lq == 0:
                etiq += 1
                lx = etiq
            elif lp != lq and lp != 0 and lq != 0:
                # Register equivalence
                T.append([lq, lp])
                lx = lq
            elif lq != 0:
                lx = lq
            elif lp != 0:
                lx = lp
            figura[i,j] = lx

# Create color image
cual = cv2.cvtColor(figura.astype(np.uint8), cv2.COLOR_GRAY2BGR)

# Draw borders of each component in a different color
for i in range(1, etiq+1):
    color = np.random.randint(0, 255, (3,))
    rows, cols = np.where(figura == i)
    for j in range(len(rows)):
        cual[rows[j],cols[j],:] = color

# Display image
cv2.imshow('Connected components', cual)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
