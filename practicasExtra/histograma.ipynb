```python


import cv2
import numpy as np
from matplotlib import pyplot as plt #pa plotear


# Lee la imagen en escala de grises
img = cv2.imread('einstein.tif', cv2.IMREAD_GRAYSCALE)

# Calcula el valor mnimo y mximo de pxeles en la imagen
min_val, max_val, _, _ = cv2.minMaxLoc(img)

# Realiza la expansin de contraste
exp_contraste = np.uint8((img - min_val) / (max_val - min_val) * 255)

# Muestra la imagen original y la imagen con expansin de contraste
# cv2.imshow('Imagen original', img)
# cv2.imshow('Expansin de contraste', exp_contraste)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

titles = ['Original Image','Expansin de contraste']
images = [img ,exp_contraste]
for i in range(2):
    plt.subplot(1,2,i+1),plt.imshow(images[i],'gray',vmin=0,vmax=255)
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python
img = cv2.imread('einstein.tif', cv2.IMREAD_GRAYSCALE)
equ = cv2.equalizeHist(img)
plt.subplot(121), plt.imshow(img, cmap='gray')
plt.title('Imagen Original'), plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(equ, cmap='gray')
plt.title('Imagen Ecualizada'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
img = cv2.imread('einstein.tif', cv2.IMREAD_GRAYSCALE)
out = cv2.normalize(img.astype('float'), None, 0.0, 1.0, cv2.NORM_MINMAX)
plt.subplot(121), plt.imshow(img, cmap='gray')
plt.title('Imagen Original'), plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(out, cmap='gray')
plt.title('Imagen Normalizada'), plt.xticks([]), plt.yticks([])

plt.show()
```
