```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar la imagen en escala de grises
img = cv2.imread('einstein.tif', cv2.IMREAD_GRAYSCALE)

# Obtener el histograma de intensidades
hist, bins = np.histogram(img.flatten(), 256, [0, 256])
plt.hist(img.flatten(), 256, [0, 256])
plt.show()
# Calcula el valor mnimo y mximo de pxeles en la imagen
min_val, max_val, _, _ = cv2.minMaxLoc(img)

# Realiza la expansin de contraste
exp_contraste = np.uint8((img - min_val) / (max_val - min_val) * 255)



titles = ['Original Image','Expansin de contraste']
images = [img ,exp_contraste]
for i in range(2):
    plt.subplot(1,2,i+1),plt.imshow(images[i],'gray',vmin=0,vmax=255)
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python


img = cv2.imread('einstein.tif', cv2.IMREAD_GRAYSCALE)
valor_blancos = 120 # Puedes cambiar este valor segn tus necesidades
valor_negros = 100 # Puedes cambiar este valor segn tus necesidades
tabla_intensidades = np.zeros((256,), dtype=np.uint8)
for i in range(256):
    if i <= valor_negros:
        tabla_intensidades[i] = 0
    elif i >= valor_blancos:
        tabla_intensidades[i] = 255
    else:
        tabla_intensidades[i] = int(255 * (i - valor_negros) / (valor_blancos - valor_negros))
out = cv2.LUT(img, tabla_intensidades)

titles = ['Original Image','Por intensidades']
images = [img ,out]
for i in range(2):
    plt.subplot(1,2,i+1),plt.imshow(images[i],'gray',vmin=0,vmax=255)
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()

```
