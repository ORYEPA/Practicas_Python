```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
x_estudio = np.random.randint(0, 20, 100)
y_nota = 1.5 + 0.3*x_estudio + np.random.randn(100)
```

```python
plt.scatter(x_estudio, y_nota, alpha=0.5, color='#53f6c7')
plt.xlabel('Horas de Estudio')
plt.ylabel('Nota Examen')
plt.title('Ejemplo Regresión Lineal')
plt.show
```

```python
from sklearn.linear_model import SGDRegressor

sgd_reg = SGDRegressor(max_iter=1000, tol=1e-3, penalty=None, 
                       learning_rate='adaptive', random_state=42)
# Cambiamos estructura de X para entrenar el modelo
x_estudio = x_estudio.reshape(-1, 1) 
sgd_reg.fit(x_estudio, y_nota)
```

```python
n = sgd_reg.intercept_


m = sgd_reg.coef_
```

```python
plt.scatter(x_estudio, y_nota, alpha=0.5, color='#53f6c7')
plt.plot(x_estudio, n + m*x_estudio, color='k')
plt.xlabel('Horas de Estudio')
plt.ylabel('Nota Examen')
plt.title('Ejemplo Regresión Lineal')
plt.show
```
